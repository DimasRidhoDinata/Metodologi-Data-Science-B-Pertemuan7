```python
#Membersihkan Applymap
universty_towns = []
with open("university_towns.txt") as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            universty_towns.append((state,line))
universty_towns[:5]
```

```python
kota_df = pd.DataFrame(universty_towns, columns=['State','Regions'])
kota_df.head(20)
```

```python
def get_citystate(item):
    if '(' in item:
        return item[:item.find('(')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
kota_df = kota_df.applymap(get_citystate)
```
